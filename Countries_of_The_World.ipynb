# Q):which country has most population density in asia?


```python
p=data.groupby('Region')['Country'].count().head(1).index.tolist()
```

```python
data_asia=data[data['Region'].isin(p)]
```

```python
data['Population_density']=data['Population_density'].str.replace(',','.')
```

```python
area=data_asia.groupby('Country')['Area'].sum()
```

```python
population=data_asia.groupby('Country')['Population'].sum()
```

```python
data_asia_population_density=(population/area).sort_values(ascending=False).reset_index()
```

```python
data_asia_population_density.head(5)
```

```python
data['Coastline']=data['Coastline'].str.replace(',','.')
data['Net migration']=data['Net migration'].str.replace(',','.')
data['Mortality']=data['Mortality'].str.replace(',','.')
data['Phones']=data['Phones'].str.replace(',','.')
data['Arable']=data['Arable'].str.replace(',','.')
data['Crops']=data['Crops'].str.replace(',','.')
data['Other']=data['Other'].str.replace(',','.')
data['Deathrate']=data['Deathrate'].str.replace(',','.')
data['Agriculture']=data['Agriculture'].str.replace(',','.')
data['Industry']=data['Industry'].str.replace(',','.')
data['Service']=data['Service'].str.replace(',','.')
data.fillna({'Climate':'0'},inplace=True)
data.fillna({'Agriculture':'0'},inplace=True)
data.fillna({'Industry':'0'},inplace=True)
data.fillna({'Service':'0'},inplace=True)
data['Birthrate']=data['Birthrate'].astype(float)
data['Coastline']=data['Coastline'].astype(float)
data['Net migration']=data['Net migration'].astype(float)
data['Mortality']=data['Mortality'].astype(float)
data['Phones']=data['Phones'].astype(float)
data['Arable']=data['Arable'].astype(float)
data['Crops']=data['Crops'].astype(float)
data['Other']=data['Other'].astype(float)
data['Deathrate']=data['Deathrate'].astype(float)
data['Agriculture']=data['Agriculture'].astype(float)
data['Industry']=data['Industry'].astype(float)
data['Service']=data['Service'].astype(float)
```
